## ZAD. 2

```python
A = ["Reszka","Orze≈Ç"]

def P(n_rzutow, n_orlow):
    from math import comb
    p = 1/2
    q = 1 - p
    return comb(n_rzutow, n_orlow) * (p**n_orlow) * (q**(n_rzutow - n_orlow))

# P(X >= 2) = P(X=2) + P(X=3)

print(P(3,2) + P(3,3))
```
